```python
import pandas as pd
import folium
```

```python
cafe_df = pd.read_csv('cafe_addresses.csv')
```

```python
cafe_df.head()
```

```python
# EU centre point
latitude, longitude = 51,10
```

```python
map_eu = folium.Map(location=[latitude, longitude], zoom_start=4)

for lat, lng, cafe, city in zip(cafe_df['Latitude'], cafe_df['Longitude'], cafe_df['Name'], cafe_df['City']):
    label = f'{cafe}, {city}'
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng], 
        radius=5,
        popup=label,
        color='#80cb34',
        weight=2,
        fill=True,
        fill_color='#CBCB34',
        fill_opacity=0.8,
        parse_html=False).add_to(map_eu)
    
map_eu
```
